```python
def getRange(start_year, end_year):
    data_for_year = weather_data[weather_data['Year'].between(
        start_year, end_year)]

    # Extract four months for which you want to calculate the averages
    # For example, let's consider January, April, July, and October
    months_of_interest = [6, 7, 8, 9]

    # Filter data for the four months
    data_for_months = data_for_year[data_for_year['Month'].isin(
        months_of_interest)]

    # data_for_months

    # Group data by month and calculate the average precipitation for each month
    average_precipitation_per_month = data_for_months.groupby(
        data_for_months['Month'])['Precipitation_Daily'].mean()

    return (average_precipitation_per_month.mean())


first = round(getRange(2021, 2040), ndigits=2)

second = round(getRange(2041, 2060), ndigits=2)
third = round(getRange(2061, 2080), ndigits=2)
fourth = round(getRange(2081, 2100), ndigits=2)

print(first)
print(second)
print(third)
print(fourth)

avg = (fourth+third+second+first)/4
print(round(avg, ndigits=2))
```
